```python
import sqlite3
import pandas as pd
```

## Carrega as bases do kaggle

```python
# le a base de de clientes
customer = pd.read_csv('../../data/raw/customer_data.csv')
customer.head(3)
```

```python
# le a base de pagamentos
payment = pd.read_csv('../../data/raw/payment_data.csv')
payment.head(3)
```

## Cria o BD credit_risk

```python
# cria um banco de dados no caminho especificado
conn = sqlite3.connect('../../data/DB/credit_risk')
cursor = conn.cursor()
```

```python
# cria a tabela de teste
cursor.execute(f'CREATE TABLE IF NOT EXISTS teste (teste)')
conn.commit()
```

```python
# cria a tabela de pagamentos
cursor.execute(f'CREATE TABLE IF NOT EXISTS payment_data {tuple(payment.columns)}')
conn.commit()
payment.to_sql('payment_data', conn, if_exists='replace', index = False)
```

```python
# cria a tabela do cliente
cursor.execute(f'CREATE TABLE IF NOT EXISTS customer_data {tuple(customer.columns)}')
conn.commit()
customer.to_sql('customer_data', conn, if_exists='replace', index = False)
```

```python
# lsita as tabelas do DB credit_risk
cursor.execute("SELECT name FROM sqlite_master WHERE type='table';")
print(cursor.fetchall())
```

```python
# deleta tabela de teste
cursor.execute(f'DROP TABLE IF EXISTS teste')
conn.commit()
```

```python
# exibe a tabela de clientes
pd.read_sql_query("select * from customer_data limit 3", conn)
```

```python
# exibe a tabela de pagamentos
pd.read_sql_query("select * from payment_data limit 3", conn)
```
